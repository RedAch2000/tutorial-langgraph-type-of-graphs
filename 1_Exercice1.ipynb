# Personalized Compliment Agent 

### 1. Importing Libraries

```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

### 2. Creating Agent State and Nodes

```python
class AgentState(TypedDict):
    name: str
    message: str = ""

def compliment_agent_node(state: AgentState) -> AgentState:
    state["message"] = state["name"] + ", you're doing an amazing job learning LangChain and LangGraph! Keep up the great work!"
    return state
```

### 3. Building the agent

```python
# creating state graph
graph = StateGraph(AgentState)

# adding the nodes
graph.add_node("compliment", compliment_agent_node)

# defining the starting point of the graph
graph.set_entry_point("compliment")

# defining the finishing point of the graph
graph.set_finish_point("compliment")

# compiling the graph
app = graph.compile()
```

```python
result = app.invoke({"name": "Reda"})
```
